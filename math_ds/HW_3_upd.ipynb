### Задание 1

Дан объект в $2D$ пространстве

````python
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]


plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

При помощи линейных отображений:
- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x (поможет матрица `[[0, 1, 0], [1, 0, 0], [0, 0, 1]]`)


````python
# Исходная матрица A (6x3)
A = np.array([
    [-100, -100, 1],
    [-100,  100, 1],
    [ 100,  100, 1],
    [   0,    0, 1],
    [ 100, -100, 1],
    [-100, -100, 1]
])

# Матрица преобразования B (3x3)
B = np.array([
    [0.5, 0,   0],
    [0,   0.5, 0],
    [0,   0,   1]
])
res = B @ A.T
res = res.T
````

````python
A = np.array(res)

x = A[:,0]
y = A[:,1]


plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

````python
angle_degrees = 130
theta = np.radians(angle_degrees)

R = np.array([
    [np.cos(theta), -np.sin(theta), 0],
    [np.sin(theta),  np.cos(theta), 0],
    [0, 0, 1]
])

res = R @ A.T
res = res.T

A = np.array(res)

x = A[:,0]
y = A[:,1]


plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

````python
R = np.array([
    [0, 1, 0],
    [1, 0, 0],
    [0, 0, 1]
])

res = R @ A.T
res = res.T

A = np.array(res)

x = A[:,0]
y = A[:,1]


plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

### Задание 2

Найдите спектральное разложение матрицы:
```
m = np.array([[1, 2],
              [2, 3]])
print(m)
```



````python
m = np.array([[1, 2],
              [2, 3]])

eigenvalues, eigenvectors = np.linalg.eig(m)
Lambda = np.diag(eigenvalues)
P = eigenvectors
reconstructed_m = P @ Lambda @ np.linalg.inv(P)

print("Lambda (собственные значения):")
print(Lambda)
print("\nP (собственные векторы):")
print(P)
print("\nВосстановленная матрица m:")
print(reconstructed_m)
````
