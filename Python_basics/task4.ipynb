```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def check_numbers(doc_number):
    for el in documents:
        if doc_number in el.values():
            return el
    return False

def find_users(doc_number):
    author = check_numbers(doc_number)
    if author:
        return author['name']
    else:
        return None

def find_directories(doc_number):
    for el in directories.items():
        if doc_number in el[1]:
            return el[0]

def select_full_info():
    for el in documents:
        pretty_string = f'№: {el["number"]}, тип: {el["type"]}, владелец: {el["name"]}, полка хранения: {find_directories(el["number"])}'
        print(pretty_string)

def find_current_directories():
    current_directories = ', '.join(list(directories.keys()))
    return current_directories

def check_directories(directorie_number):
    return directorie_number in directories

def check_empty_dir(directorie_number):
    return directories[directorie_number] == []

def add_directories(directorie_number):
    directories.setdefault(directorie_number,[])

def delete_directories(directorie_number):
    del directories[directorie_number]

def handle_find_user():
    doc_number = input('Введите номер документа:')
    user = find_users(doc_number)
    print(f'Владелец документа: {user}' if user != None else f'Документ не найден в базе')

def handle_find_directory():
    doc_number = input('Введите номер документа:')
    dir = find_directories(doc_number)
    print(f'Документ хранится на полке: {dir}' if dir != None else f'Документ не найден в базе')


def handle_add_directory():
    directorie_number = input('Введите номер полки:')
    if check_directories(directorie_number):
        print(f'Такая полка уже существует. Текущий перечень полок: {find_current_directories()}.')
    else:
        add_directories(directorie_number)
        print(f'Полка добавлена. Текущий перечень полок: {find_current_directories()}.')

def handle_delete_directory():
    directorie_number = input('Введите номер полки:').strip()

    if not check_directories(directorie_number):
        print(f'Такой полки не существует. Текущий перечень полок: {find_current_directories()}.')
    elif not check_empty_dir(directorie_number):
        print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {find_current_directories()}.')
    else:
        delete_directories(directorie_number)
        print(f'Полка удалена. Текущий перечень полок: {find_current_directories()}.')

#функции для второго задания

def add_directory(dir,number):
    if check_directories(dir):
        if dir in directories.keys():
            directories[dir].append(number)
        else:
            directories.setdefault(dir,[number])
    else: return False

def add_document(number, doc_type, owner):
    documents.append({'type':doc_type,'number':number,'name':owner})

def delete_document(el):
    documents.remove(el)

def delete_el_in_dir(number):
    dir = find_directories(number)
    directories[dir].remove(number)

def transfer_document(number,dir_new):
    delete_el_in_dir(number)
    directories[dir_new].append(number)

def handle_add_document():
    number, doc_type, owner, directory = (
            input('Введите номер документа: '),
            input('Введите тип документа: '),
            input('Введите владельца документа: '),
            input('Введите полку для хранения: ')
        )
    if check_directories(directory):
        add_document(number, doc_type, owner)
        add_directory(directory,number)
        print('Текущий список документов:')
        select_full_info()
    else:
        print(f'Такой полки не существует. Добавьте полку командой ads\nТекущий список документов:')
        select_full_info()

def handle_delete_document():
    number = input('Введите номер документа: ')
    el = check_numbers(number)
    if el:
        delete_document(el)
        delete_el_in_dir(number)
        print(f'Документ удален\nТекущий список документов:')
        select_full_info()
    else:
        print(f'Документ не найден в базе\nТекущий список документов:')
        select_full_info()

def handle_transfer_document():
    number, dir = (
        input('Введите номер документа: '),
        input('Введите номер полки: '))
    if check_directories(dir) and check_numbers(number):
       transfer_document(number,dir)
       print(f'Документ перемещен\nТекущий список документов:')
       select_full_info()
    elif not check_directories(dir):
        print(f'Такой полки не существует. Текущий перечень полок: {find_current_directories()}.')
    else:
        print(f'Документ не найден в базе\nТекущий список документов:')
        select_full_info()


# main для google colab

def main():
    commands = {
        'p': handle_find_user, # найти владельца документа
        's': handle_find_directory, #найти документ на полке
        'l': select_full_info, #получить полную информацию по документам
        'ads': handle_add_directory, #добавить новую полку
        'ds': handle_delete_directory, #удалить полку
        'ad': handle_add_document, #добавить документ в данные
        'd': handle_delete_document, # удалить документ из данных
        'm': handle_transfer_document, #переместить документ с полки на полку
        'q': "q"
    }
    while True:
        command = input('Введите команду: ').strip()
        action = commands.get(command)

        if action != 'q':
            action()
        elif action == 'q':
          break
        else:
            print('Команда не найдена')
main()

# main для pycharm - почему-то quit() или exit() не работает в google colab

# def main():
#     commands = {
#         'p': handle_find_user, # найти владельца документа
#         's': handle_find_directory, #найти документ на полке
#         'l': select_full_info, #получить полную информацию по документам
#         'ads': handle_add_directory, #добавить новую полку
#         'ds': handle_delete_directory, #удалить полку
#         'ad': handle_add_document, #добавить документ в данные
#         'd': handle_delete_document, # удалить документ из данных
#         'm': handle_transfer_document, #переместить документ с полки на полку
#         'q': quit
#     }

#     while True:
#         command = input('Введите команду: ').strip()
#         action = commands.get(command)

#         if action:
#             action()
#         else:
#             print('Команда не найдена')
# main()
```
